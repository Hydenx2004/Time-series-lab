```python
import pandas as pd
```

```python
df = pd.read_csv('dailymintemp.csv', parse_dates=['Date'],index_col = 'Date')
```

```python
df.head()
```

```python
df['day']=df.index.day
df['month']=df.index.month
df['year']=df.index.year
df['isweekend']=df.index.dayofweek >=5
```

```python
df.head()
```

```python
df['lag1'] = df['Temp'].shift(1)
df['lag2'] = df['Temp'].shift(2)
```

```python
df.head()
```

```python
df['rolling_mean'] = df['Temp'].rolling(3).mean()
```

```python
df['rolling_std'] = df['Temp'].rolling(3).std()
```

```python
df['expanding_mean'] = df['Temp'].expanding(3).mean()
```

```python
df['expanding_max'] = df['Temp'].expanding(3).max()
```
