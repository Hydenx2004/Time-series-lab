```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dailymintemp.csv", parse_dates=['Date'], index_col='Date')
```

```python
df.head()
```

```python
plt.figure(figsize=(8,4))
plt.plot(df.index, df['Temp'])
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Daily Minimum Temperatures')
plt.show()
```

```python
plt.figure(figsize=(8,3))
pd.plotting.autocorrelation_plot(df['Temp'])
plt.title("Autocorrelation plot")
plt.show()
```

```python
plt.figure(figsize=(8,3))
plt.hist(df['Temp'], bins=10)
plt.title("Histogram for Daily Temp in Jan24")
plt.xlabel("Temperature in Celsius")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,3))
sns.kdeplot(df['Temp'])
plt.title('Density plot for Daily Temp in Jan24')
plt.xlabel('Temperature')
plt.show()
```

```python
plt.figure(figsize=(2,2))
sns.boxplot(x=df['Temp'])
plt.title('Box and Whixker plot for Daily Temp in Jan24')
plt.xlabel('Temperature')
plt.show()
```

```python
plt.figure(figsize=(2,2))
sns.heatmap(df, cmap='coolwarm')
plt.title('Heatmap for Daily Temp in Jan24')
plt.xlabel('Temperature')
plt.show()
```
